```python
#####____Data Cleaning____#####
'''
Bad data could be:
* Empty cells
* Data in wrong format
* Wrong data
* Duplicates

'''

df_dirty = pd.read_csv('/content/drive/MyDrive/dirtydata.csv')
print(df_dirty.to_string())
```

```python
#Removing rows having empty cell.

#dropna(inplace = True) method: removes the empty cell and change the original dataframe because inplace = True.

new_df_dirty = df_dirty.dropna()
print(new_df_dirty)

#Row 22 was removed now and original data remain as it is.
```

```python
#Replacing the empty cell/Null values with new value, instead of deleting it.

#fillna(value, inplace = True) method allows us to replace empty cells with a value.
df_dirty_1 = df_dirty
print(df_dirty_1)
```

```python
df_dirty_1.fillna(120, inplace=True)

print(df_dirty_1)
```

```python
#Replacing only the specified columns.

df_dirty_2 = pd.read_csv('/content/drive/MyDrive/dirtydata.csv')
print(df_dirty_2)
```

```python
df_dirty_2['Calories'].fillna(000, inplace=True)
print(df_dirty_2)
```

```python
#Replacing the empty cell by mean().

df_dirty_2 = pd.read_csv('/content/drive/MyDrive/dirtydata.csv')
print(df_dirty_2)
```

```python
x = df_dirty_2['Calories'].mean()
print(x)
```

```python
df_dirty_2['Calories'].fillna(x, inplace=True)
print(df_dirty_2)
```

```python
#Replacing the empty cell by median():The value in the middle, after you have sorted all values ascending.

df_dirty_2 = pd.read_csv('/content/drive/MyDrive/dirtydata.csv')
print(df_dirty_2, '\n')

x = df_dirty_2['Calories'].median()
print(x)
```

```python
df_dirty_2['Calories'].fillna(x, inplace=True)
print(df_dirty_2)
```

```python
#Replacing with mode():the value that appears most frequently.
df_dirty_2 = pd.read_csv('/content/drive/MyDrive/dirtydata.csv')
print(df_dirty_2, '\n')

x = df_dirty_2['Calories'].mode()[0]
print(x)
```

```python
df_dirty_2['Calories'].fillna(x, inplace=True)
print(df_dirty_2)
```
